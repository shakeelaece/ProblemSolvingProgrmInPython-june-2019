```python
# Linear Graph

import matplotlib.pyplot as plt

plt.title('Basic Linear Graph')
plt.plot([1,2,3],[1,2,3])   
plt.show()
```

```python
# Graph to plot distance and time

distance=list(range(2,11,2))
time=list(range(1,6))
#d=2*t  --> this is our condition for time & distance

plt.title("Distance vs Time")
plt.plot(time,distance,'go')
plt.show()
```

```python
# Graph to plot distance and time

distance=list(range(2,11,2))
time=list(range(1,6))
#d=2*t  --> this is our condition for time & distance

plt.title("Distance vs Time")
plt.xlabel('Time')
plt.ylabel("Distance")
plt.plot(time,distance,'bo')
plt.grid()
plt.show()
```

```python
# Non-Linear Graph
# Ex: v=a*t

t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t)) ]

plt.title("Velocity vs Time")
plt.xlabel("Time")
plt.ylabel("Velocity")
plt.plot(t,v,'yo')
plt.grid()
plt.show()
```

```python
# Graph for Numbers vs cubes

n=list(range(10))
sqn=[i**3 for i in n]

plt.title("Exponential Graph")
plt.plot(n,sqn,'r')
plt.grid()
plt.savefig('Images/Exponential.png')
plt.show()
```

```python
# Sinusoidal Graph for Voltage-Time

import numpy as np

#t=[i/10 for i in range(1,21)]  OR
t=np.arange(0.1,2.0,0.1)
v=[1+np.sin(2*np.pi*i) for i in t]
plt.plot(t,v,'b*')
plt.grid()
plt.xlabel("Time")
plt.ylabel("Voltage")
plt.title("Voltage vs Time")
plt.savefig('Images/Voltage-Time.png')
plt.show()
```

```python
# plot Income of Arkansas from 2005 to 2013

import pandas as pd
df=pd.read_csv('DataFiles/income.csv')
years=df.columns[2:]

incomeArkansas=df.values[3,2:]   # Here 3 is row index of Arkansas in income file

plt.title("Income of Arkansas")
plt.xlabel('Year')
plt.ylabel('Income')
plt.plot(years,incomeArkansas)
plt.show()
```

```python
# Income of all states from 2005 to 2013

df
incomeallstates=[]

for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
plt.plot(years,incomeallstates)
plt.title("Income of All States")
plt.xlabel("Years")
plt.ylabel("Income")
plt.grid()
plt.show()
```
